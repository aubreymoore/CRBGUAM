# Trap Catch Summary

```python
%%sql
SELECT trap.trap_id, latitude, longitude, location_description, 
    MIN(visit_date), MAX(visit_date), COUNT(visit_date) AS visits,
    SUM(male_count+female_count+unsexed_count) AS beetlesTrapped
FROM trap, trap_visit
WHERE trap.trap_id = trap_visit.trap_id
    AND trap.trap_id IN ('x1963', 'x1975', 'x1974', 'x1976', 'x1969', 'x1962', 'x1971')
GROUP BY trap.trap_id
ORDER BY trap.trap_id
```
