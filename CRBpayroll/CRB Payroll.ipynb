# Run the Simulation

```python
# Create a series for grant personnel cost balances
balance = grantsdf['personnelTotal']

# Create a series of pay days (weekly)
paydays = pandas.date_range(start='20140502', end='20150809', freq='W')

benefits = 0.3149

def payout(row):
    name = row['name']
    grant = row['grant']
    # Pay the staff member
    hourly = staffdf.loc[name, 'hourly']
    maxHours = staffdf.loc[name, 'maxHours']
    pay = (1.0 + benefits) * hourly * maxHours
    balance[grant] -= pay
    outf.write(paydaystr + ",'" + name + "'," + "%.2f" %pay + "," +  grant + "," + "%.2f" % balance[grant] + "\n")
    
outf = open('simresults.csv', 'w')
outf.write('date,name,pay,grant,balance\n')

for payday in paydays:
    paydaystr = payday.strftime('%Y%m%d')
    mask1 = int(paydaystr) >= paysourcedf.start
    mask2 = int(paydaystr) <= paysourcedf.end
    x = paysourcedf[mask1 & mask2]
    x.apply(payout, axis=1)  
outf.close()
```
